```python
import pandas as pd
```

```python
confirmed_df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
```

```python
confirmed_df.head()
```

```python
list(confirmed_df.columns[4: ])
```

```python
confirmed_processed = pd.melt(confirmed_df, id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], value_vars=list(confirmed_df.columns[4: ]), var_name='date', value_name='confirmed')
```

```python
confirmed_processed.head()
```

```python
deaths_df = pd.read_csv('https://raw.github.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
recovered_df = pd.read_csv('https://raw.github.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')
```

```python
deaths_processed = pd.melt(deaths_df, id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], value_vars=list(deaths_df.columns[4: ]), var_name='date', value_name='deaths')
```

```python
deaths_processed
```

```python
recovered_processed = pd.melt(recovered_df, id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], value_vars=list(recovered_df.columns[4: ]), var_name='date', value_name='recovered')
```

```python
recovered_processed
```

```python
from functools import reduce
```

```python
covid_data = [confirmed_processed, deaths_processed, recovered_processed]
```

```python
covid_merged = reduce(lambda left, right: pd.merge(left, right, on=['Province/State', 'Country/Region', 'Lat', 'Long', 'date'], how='outer'), covid_data).fillna(0)
```
